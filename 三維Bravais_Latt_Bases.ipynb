# **簡單三斜 (Simple Triclinic)**
a≠b≠c，α≠β≠γ≠90°

```python
# 1. Import necessary libraries
import lattpy as lp
import matplotlib.pyplot as plt
import numpy as np


def main():

    # 2. Define the Monoclinic lattice parameters
    # Lattice constants (a ≠ b ≠ c)
    a = 2.0
    b = 2.5
    c = 3.0


    # Interaxial angle in degrees (α = γ = 90°, β ≠ 90°)
    alpha_deg = 80
    beta_deg = 65
    gamma_deg = 70


    # 3. Convert parameters to Cartesian vectors (Your original, correct method)
    # First, convert the non-90 angle from degrees to radians for numpy's trig functions
    beta_rad = np.deg2rad(beta_deg)

    # Define the basis vectors in Cartesian coordinates.
    # This choice sets 'a' along x, 'b' along y, and 'c' in the xz-plane.
    v1 = [a, 0, 0]
    v2 = [0, b, 0]
    v3 = [c * np.cos(beta_rad), 0, c * np.sin(beta_rad)]

    # Combine vectors into a single matrix that lp.Lattice() understands
    vectors = np.array([v1, v2, v3])

    # 4. Create the lattice object using the generated vectors
    monoclinic_lattice = lp.Lattice(vectors)


    # 5. Plot the lattice unit cell
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    monoclinic_lattice.plot_cell(ax=ax, show=False,  lw=2)

    # IMPORTANT: Adjust the viewing angle to see the 3D structure
    ax.view_init(elev=22, azim=40)

    ax.set_title("Monoclinic Lattice Unit Cell with Parameters", fontsize=14)

    # Get the basis vectors from the lattice object
    basis_vectors = monoclinic_lattice.vectors

    # --- Annotate Edge Lengths (a, b, c) ---
    # Vector 1 (a)
    ax.text(basis_vectors[0, 0] * 0.5, basis_vectors[0, 1] * 0.5, basis_vectors[0, 2] * 0.5,
            f'a={a}', color='red', fontsize=12, ha='center', va='center')
    # Vector 2 (b)
    ax.text(basis_vectors[1, 0] * 0.5, basis_vectors[1, 1] * 0.5, basis_vectors[1, 2] * 0.5,
            f'b={b}', color='green', fontsize=12, ha='center', va='center')
    # Vector 3 (c)
    ax.text(basis_vectors[2, 0] * 0.5, basis_vectors[2, 1] * 0.5, basis_vectors[2, 2] * 0.5,
            f'c={c}', color='blue', fontsize=12, ha='center', va='center')


    # --- Annotate Angles (alpha, beta, gamma) ---
    # We place these slightly offset from the origin for visibility

    # Alpha (angle between b and c vectors)
    # Since lattpy's monoclinic aligns 'a' with x-axis, 'b' with y-axis,
    # and 'c' in xz-plane, the angle alpha (between b and c) is 90 degrees.
    # We'll place it near the origin, slightly in the b-c plane's positive quadrant.
    ax.text(0.1, 0.5, 0.5, f'α={alpha_deg}°', color='purple', fontsize=12, ha='center', va='center')

    # Beta (angle between a and c vectors)
    # This is the non-90 degree angle for monoclinic
    # Place it near the origin, slightly in the a-c plane's positive quadrant.
    ax.text(0.5, 0.1, 0.5, f'β={beta_deg}°', color='darkorange', fontsize=12, ha='center', va='center')

    # Gamma (angle between a and b vectors)
    # This is 90 degrees for monoclinic
    # Place it near the origin, slightly in the a-b plane's positive quadrant.
    ax.text(0.5, 0.5, 0.1, f'γ={gamma_deg}°', color='darkgreen', fontsize=12, ha='center', va='center')


    # Set labels for axes
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')

    # Set limits to make sure all annotations are visible and centered
    max_dim = max(a,b,c) * 1.2
    ax.set_xlim([-0.1, max_dim])
    ax.set_ylim([-0.1, max_dim])
    ax.set_zlim([-0.1, max_dim])

    plt.show()

    # You can also print the object to verify the vectors
    print("--- Lattice Information ---")
    print(monoclinic_lattice)


if __name__ == "__main__":
    main()
```
